# Exploratory Data Analysis - User Data

This notebook provides an exploratory data analysis of the user survey data from Rwanda.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load the data
df = pd.read_csv('user_data.csv')

print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Basic information about the dataset
print("Dataset Info:")
print("=" * 60)
df.info()

print("\n\nMissing Values:")
print("=" * 60)
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percentage': missing_pct
})
missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
```

```python
# Summary statistics for numerical variables
print("Summary Statistics for Numerical Variables:")
print("=" * 60)
df.describe()
```

## 1. Demographics Analysis


```python
# Gender distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
gender_counts = df['gender'].value_counts()
axes[0].bar(gender_counts.index, gender_counts.values, color=['#FF6B9D', '#4ECDC4'])
axes[0].set_title('Gender Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')
for i, v in enumerate(gender_counts.values):
    axes[0].text(i, v + 1000, f'{v:,}\n({v/len(df)*100:.1f}%)', 
                 ha='center', va='bottom', fontweight='bold')

# Pie chart
axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%',
           colors=['#FF6B9D', '#4ECDC4'], startangle=90)
axes[1].set_title('Gender Distribution (Percentage)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"Gender Distribution:")
print(gender_counts)
print(f"\nPercentage:")
print((gender_counts / len(df) * 100).round(2))
```

```python
# Age distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df['age'], bins=13, color='#95E1D3', edgecolor='black', alpha=0.7)
axes[0].set_title('Age Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].axvline(df['age'].mean(), color='red', linestyle='--', linewidth=2, 
                label=f'Mean: {df["age"].mean():.1f}')
axes[0].axvline(df['age'].median(), color='blue', linestyle='--', linewidth=2, 
                label=f'Median: {df["age"].median():.1f}')
axes[0].legend()

# Box plot by gender
df.boxplot(column='age', by='gender', ax=axes[1], grid=False)
axes[1].set_title('Age Distribution by Gender', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Age')
plt.suptitle('')  # Remove default title

plt.tight_layout()
plt.show()

print(f"Age Statistics:")
print(f"Mean: {df['age'].mean():.2f}")
print(f"Median: {df['age'].median():.2f}")
print(f"Std: {df['age'].std():.2f}")
print(f"Min: {df['age'].min()}")
print(f"Max: {df['age'].max()}")
```

## 2. Education and Employment


```python
# Education level distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Count plot
edu_counts = df['education_level'].value_counts()
edu_order = ['below primary', 'primary', 'secondary', 'secondary-professional', 
             'professional', 'tertiary-and-above']
edu_counts = df['education_level'].value_counts().reindex(edu_order, fill_value=0)

axes[0].barh(range(len(edu_counts)), edu_counts.values, color='#F38181')
axes[0].set_yticks(range(len(edu_counts)))
axes[0].set_yticklabels(edu_counts.index)
axes[0].set_title('Education Level Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Count')
for i, v in enumerate(edu_counts.values):
    axes[0].text(v + 500, i, f'{v:,} ({v/len(df)*100:.1f}%)', 
                 va='center', fontweight='bold')

# Employment status
employment_data = pd.DataFrame({
    'Has Occupation': df['occupation'].value_counts(),
    'Informal Working': df['informal_working'].value_counts()
})

axes[1].bar(['No', 'Yes'], [df[df['occupation']==0].shape[0], df[df['occupation']==1].shape[0]], 
           color='#AA96DA', alpha=0.7, label='Has Occupation')
axes[1].bar(['No', 'Yes'], [df[df['informal_working']==0].shape[0], df[df['informal_working']==1].shape[0]], 
           color='#FCBAD3', alpha=0.7, label='Informal Working')
axes[1].set_title('Employment Status', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Count')
axes[1].legend()

plt.tight_layout()
plt.show()

print("Education Level Distribution:")
print((df['education_level'].value_counts() / len(df) * 100).round(2))
print(f"\nEmployment:")
print(f"Has Occupation: {df['occupation'].sum():,} ({df['occupation'].mean()*100:.1f}%)")
print(f"Informal Working: {df['informal_working'].sum():,} ({df['informal_working'].mean()*100:.1f}%)")
```

## 3. Disability and Marriage Status


```python
# Disability and marriage status
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Disability
disability_counts = df['disability'].value_counts()
axes[0].bar(['No Disability', 'Has Disability'], disability_counts.values, 
           color=['#95E1D3', '#F38181'])
axes[0].set_title('Disability Status', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(disability_counts.values):
    axes[0].text(i, v + 1000, f'{v:,}\n({v/len(df)*100:.1f}%)', 
                 ha='center', va='bottom', fontweight='bold')

# Marriage status
marriage_counts = df['marriage_status'].value_counts()
axes[1].bar(['Not Married', 'Married'], marriage_counts.values, 
           color=['#AA96DA', '#FCBAD3'])
axes[1].set_title('Marriage Status', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Count')
for i, v in enumerate(marriage_counts.values):
    axes[1].text(i, v + 1000, f'{v:,}\n({v/len(df)*100:.1f}%)', 
                 ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 4. Livestock and Assets


```python
# Livestock ownership
livestock_cols = ['num_cows', 'num_goats', 'num_chickens', 'num_sheep', 'num_pigs', 'num_rabbits']
livestock_data = df[livestock_cols].sum().sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Total livestock counts
axes[0].barh(livestock_data.index, livestock_data.values, color='#95E1D3')
axes[0].set_title('Total Livestock Counts', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Total Count')
for i, v in enumerate(livestock_data.values):
    axes[0].text(v + 500, i, f'{v:,}', va='center', fontweight='bold')

# Percentage of households with each type
has_livestock = (df[livestock_cols] > 0).sum() / len(df) * 100
axes[1].barh(has_livestock.index, has_livestock.values, color='#F38181')
axes[1].set_title('Percentage of Households with Livestock', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Percentage (%)')
for i, v in enumerate(has_livestock.values):
    axes[1].text(v + 1, i, f'{v:.1f}%', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("Livestock Statistics:")
print(df[livestock_cols].describe())
```

```python
# Land ownership
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Land ownership
land_own_counts = df['land_ownership'].value_counts()
axes[0].bar(['No Land', 'Has Land'], land_own_counts.values, color=['#AA96DA', '#FCBAD3'])
axes[0].set_title('Land Ownership', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(land_own_counts.values):
    axes[0].text(i, v + 1000, f'{v:,}\n({v/len(df)*100:.1f}%)', 
                 ha='center', va='bottom', fontweight='bold')

# Land size distribution (for those who own land)
land_owners = df[df['land_ownership'] == 1]['land_size'].dropna()
if len(land_owners) > 0:
    axes[1].hist(land_owners, bins=20, color='#95E1D3', edgecolor='black', alpha=0.7)
    axes[1].set_title('Land Size Distribution (Hectares)', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Land Size (Hectares)')
    axes[1].set_ylabel('Frequency')
    axes[1].axvline(land_owners.mean(), color='red', linestyle='--', linewidth=2, 
                    label=f'Mean: {land_owners.mean():.2f}')
    axes[1].legend()
else:
    axes[1].text(0.5, 0.5, 'No land owners', ha='center', va='center', fontsize=14)

plt.tight_layout()
plt.show()

print(f"Land Ownership: {df['land_ownership'].sum():,} ({df['land_ownership'].mean()*100:.1f}%)")
if len(land_owners) > 0:
    print(f"\nLand Size Statistics (for owners):")
    print(f"Mean: {land_owners.mean():.2f} hectares")
    print(f"Median: {land_owners.median():.2f} hectares")
    print(f"Min: {land_owners.min():.2f} hectares")
    print(f"Max: {land_owners.max():.2f} hectares")
```

## 5. Electronics and Technology


```python
# Electronics ownership
electronics_cols = ['num_radio', 'num_phone', 'num_tv']
electronics_data = df[electronics_cols]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Total counts
total_electronics = electronics_data.sum()
axes[0].bar(total_electronics.index, total_electronics.values, color='#4ECDC4')
axes[0].set_title('Total Electronics Counts', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Total Count')
for i, v in enumerate(total_electronics.values):
    axes[0].text(i, v + 2000, f'{v:,}', ha='center', va='bottom', fontweight='bold')

# Percentage with at least one
has_electronics = (electronics_data > 0).sum() / len(df) * 100
axes[1].bar(has_electronics.index, has_electronics.values, color='#FF6B9D')
axes[1].set_title('Percentage of Households with Electronics', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Percentage (%)')
axes[1].set_ylim(0, 100)
for i, v in enumerate(has_electronics.values):
    axes[1].text(i, v + 2, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("Electronics Statistics:")
print(df[electronics_cols].describe())
print(f"\nPercentage with at least one:")
for col in electronics_cols:
    pct = (df[col] > 0).sum() / len(df) * 100
    print(f"{col}: {pct:.1f}%")
```

## 6. Housing and Infrastructure


```python
# Housing quality indicators
housing_cols = ['floor', 'roof', 'walls', 'toilet']
housing_data = df[housing_cols]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, col in enumerate(housing_cols):
    counts = df[col].value_counts()
    axes[idx].bar(['Basic', 'Improved'], [counts.get(0, 0), counts.get(1, 0)], 
                 color=['#F38181', '#95E1D3'])
    axes[idx].set_title(f'{col.capitalize()} Quality', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Count')
    total = counts.sum()
    for i, (key, val) in enumerate([(0, counts.get(0, 0)), (1, counts.get(1, 0))]):
        pct = val / total * 100
        axes[idx].text(i, val + 1000, f'{val:,}\n({pct:.1f}%)', 
                      ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("Housing Quality Statistics:")
for col in housing_cols:
    improved_pct = (df[col] == 1).sum() / len(df) * 100
    print(f"{col.capitalize()}: {improved_pct:.1f}% have improved quality")
```

## 7. Fuel and Water Source


```python
# Fuel and water source distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Fuel types
fuel_counts = df['fuel'].value_counts()
axes[0].bar(fuel_counts.index, fuel_counts.values, color='#AA96DA')
axes[0].set_title('Fuel Type Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Fuel Type')
axes[0].set_ylabel('Count')
for i, v in enumerate(fuel_counts.values):
    axes[0].text(i, v + 1000, f'{v:,}\n({v/len(df)*100:.1f}%)', 
                 ha='center', va='bottom', fontweight='bold')

# Water sources
water_counts = df['water_source'].value_counts()
axes[1].bar(water_counts.index, water_counts.values, color='#4ECDC4')
axes[1].set_title('Water Source Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Water Source')
axes[1].set_ylabel('Count')
for i, v in enumerate(water_counts.values):
    axes[1].text(i, v + 1000, f'{v:,}\n({v/len(df)*100:.1f}%)', 
                 ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("Fuel Type Distribution:")
print((df['fuel'].value_counts() / len(df) * 100).round(2))
print(f"\nWater Source Distribution:")
print((df['water_source'].value_counts() / len(df) * 100).round(2))
```

## 8. Cross-tabulations and Relationships


```python
# Education by Gender
edu_gender = pd.crosstab(df['education_level'], df['gender'], normalize='columns') * 100
edu_order = ['below primary', 'primary', 'secondary', 'secondary-professional', 
             'professional', 'tertiary-and-above']
edu_gender = edu_gender.reindex(edu_order)

fig, ax = plt.subplots(figsize=(12, 6))
edu_gender.plot(kind='bar', ax=ax, color=['#FF6B9D', '#4ECDC4'])
ax.set_title('Education Level Distribution by Gender', fontsize=14, fontweight='bold')
ax.set_xlabel('Education Level')
ax.set_ylabel('Percentage (%)')
ax.set_xticklabels(edu_gender.index, rotation=45, ha='right')
ax.legend(title='Gender')
plt.tight_layout()
plt.show()

print("Education Level by Gender (%):")
print(edu_gender.round(2))
```

```python
# Age distribution by education level
fig, ax = plt.subplots(figsize=(14, 6))
edu_order = ['below primary', 'primary', 'secondary', 'secondary-professional', 
             'professional', 'tertiary-and-above']
df['education_level_ordered'] = pd.Categorical(df['education_level'], categories=edu_order, ordered=True)

df.boxplot(column='age', by='education_level_ordered', ax=ax, grid=False)
ax.set_title('Age Distribution by Education Level', fontsize=14, fontweight='bold')
ax.set_xlabel('Education Level')
ax.set_ylabel('Age')
plt.suptitle('')  # Remove default title
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
